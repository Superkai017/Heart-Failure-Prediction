```python
col = list(df.columns)
categorical_features = []
numerical_features = []
for i in col:
    if len(df[i].unique()) > 6:
        numerical_features.append(i)
    else:
        categorical_features.append(i)

print('Categorical Features :',*categorical_features)
print('Numerical Features :',*numerical_features)
```

```python
ncols = 2
nrows = int(np.ceil(len(categorical_features) / ncols))

fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 5 * nrows))
ax = ax.flatten()

for i, feature in enumerate(categorical_features):
    sns.countplot(data=df, x=feature, ax=ax[i], palette='Set2')
    ax[i].set_title(f'Distribution of {feature}')
    ax[i].set_xlabel(feature)
    ax[i].set_ylabel('Count')

for j in range(len(categorical_features), len(ax)):
    fig.delaxes(ax[j])

plt.tight_layout()
plt.show()
```

```python
fig , ax = plt.subplots(nrows=3, ncols=2, figsize=(20, 15))
ax = ax.flatten()
for i, feature in enumerate(numerical_features):
    sns.histplot(data=df, x=feature, ax=ax[i], kde=True, color='navy')
    ax[i].set_title(f'Distribution of {feature}')
    ax[i].set_xlabel(feature)
    ax[i].set_ylabel('Frequency')
for j in range(len(numerical_features), len(ax)):
    fig.delaxes(ax[j])
plt.tight_layout()
plt.show()
```

```python
import warnings
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import precision_recall_curve
warnings.filterwarnings('ignore')
```

```python

colors = 'coolwarm'
from sklearn.metrics import RocCurveDisplay
def model(classifier):
    X = df2.drop('HeartDisease', axis=1)
    y = df2['HeartDisease']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
    classifier.fit(X_train, y_train)
    prediction = classifier.predict(X_test)
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    print("Accuracy : ", '{0:.2%}'.format(accuracy_score(y_test, prediction)))
    print("Cross Validation Score : ", '{0:.2%}'.format(cross_val_score(classifier, X_train, y_train, cv=cv, scoring='roc_auc').mean()))
    print("ROC_AUC Score : ", '{0:.2%}'.format(roc_auc_score(y_test, prediction)))
    RocCurveDisplay.from_estimator(classifier, X_test, y_test)
    plt.title('ROC_AUC_Plot')
    plt.show()

def model_evaluation(classifier):
    X = df2.drop('HeartDisease', axis=1)
    y = df2['HeartDisease']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
    # Confusion Matrix
    cm = confusion_matrix(y_test, classifier.predict(X_test))
    names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
    counts = [value for value in cm.flatten()]
    percentages = ['{0:.2%}'.format(value) for value in cm.flatten()/np.sum(cm)]
    labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in zip(names, counts, percentages)]
    labels = np.asarray(labels).reshape(2, 2)
    sns.heatmap(cm, annot=labels, cmap=colors, fmt='')
    
    # Classification Report
    print(classification_report(y_test, classifier.predict(X_test)))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 0, C = 10 , penalty = 'l2')
model(lr)
```

```python
model_evaluation(lr)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 0, n_estimators = 100, max_depth = 5)
model(rf)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
model(knn)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf', random_state=0, C=1)
model(svc)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=0, max_depth=5)
model(dt)
```

```python
model_evaluation(dt)
```
