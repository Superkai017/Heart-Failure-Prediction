```python
num_cols = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak', 'HeartDisease']
```

```python

df['RestingBP'] = df['RestingBP'].replace(0, np.nan)
df['RestingBP'] = df['RestingBP'].fillna(df['RestingBP'].median())

df['Cholesterol'] = df['Cholesterol'].replace(0, np.nan)
df['Cholesterol'] = df['Cholesterol'].fillna(df['Cholesterol'].median())
```

```python
df['Oldpeak'] = df['Oldpeak'].apply(lambda x: np.nan if x < 0 else x)
df['Oldpeak'] = df['Oldpeak'].fillna(df['Oldpeak'].median())
```

```python
df.describe()
```

Encoding


```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])             # M/F → 1/0
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
```

```python
df = pd.get_dummies(df, columns=['ChestPainType', 'RestingECG', 'ST_Slope'])
```

```python
df.dtypes
```

```python
df.head()
```

Featuring 

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
mms = MinMaxScaler() # Normalization
ss = StandardScaler() # Standardization

df['Oldpeak'] = mms.fit_transform(df[['Oldpeak']])
df['Age'] = ss.fit_transform(df[['Age']])
df['RestingBP'] = ss.fit_transform(df[['RestingBP']])
df['Cholesterol'] = ss.fit_transform(df[['Cholesterol']])
df['MaxHR'] = ss.fit_transform(df[['MaxHR']])
df.head()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif, chi2
X = df.drop('HeartDisease', axis=1)
y = df['HeartDisease']
anova_selector = SelectKBest(score_func=f_classif, k='all')
anova_selector.fit(X, y)

anova_df = pd.DataFrame({
    'Feature': X.columns,
    'F_score': anova_selector.scores_,
    'p_value': anova_selector.pvalues_
}).sort_values('F_score', ascending=False)

print("=== ANOVA F-test ===")
print(anova_df.to_string(index=False))

# ── Chi-Square test (requires non-negative values) ────────────────────────────
# Scale to [0,1] since data is z-score normalized (has negatives)
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

chi2_selector = SelectKBest(score_func=chi2, k='all')
chi2_selector.fit(X_scaled, y)

chi2_df = pd.DataFrame({
    'Feature': X.columns,
    'Chi2_score': chi2_selector.scores_,
    'p_value': chi2_selector.pvalues_
}).sort_values('Chi2_score', ascending=False)

print("\n=== Chi-Square test ===")
print(chi2_df.to_string(index=False))

# ── Drop features where BOTH tests give p_value > 0.05 ───────────────────────
anova_drop = set(anova_df[anova_df['p_value'] > 0.05]['Feature'])
chi2_drop  = set(chi2_df[chi2_df['p_value']  > 0.05]['Feature'])
cols_to_drop = anova_drop & chi2_drop  # only drop if BOTH agree

df_selected = df.drop(columns=list(cols_to_drop))

print(f"\n=== Dropped (failed both tests): {list(cols_to_drop)} ===")
print(f"Remaining shape: {df_selected.shape}")
print(f"Kept features: {df_selected.columns.tolist()}")
```

```python
cols_to_drop = ['ChestPainType_TA', 'RestingECG_LVH']

df_selected = df.drop(columns=cols_to_drop)

print("Shape before:", df.shape)
print("Shape after: ", df_selected.shape)
print("Dropped:     ", cols_to_drop)
```
